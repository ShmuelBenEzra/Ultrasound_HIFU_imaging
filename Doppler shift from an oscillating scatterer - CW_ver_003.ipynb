# Extracting the Doppler shift from an oscillating scatterer

## Preface: the Doppler shift from a scatterer moving at constant velocity - the CW analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as fft
from scipy import signal
```

```python
# Model assumptions
c = 1500.0 # [m/s] spped of sound
Z = 1.5e+6 # [Ray.] mechanical impedance
Fs = 50.0e+6 # [Hz] sampling frequency
f = 6.0e+6 # [Hz] carrier frequency
u = 1.0 # [m/s] the (constant) velocity of the scatterer (parallel component, i.e. theta=0)
# scattering coefficient = 1.0
_2Pi_ = 2*np.pi
_2Pi_
```

```python
# Doppler shift
fd = 2.0*u/c*f
f2 = f + fd
fd
```

```python
DF = 200.0 #[Hz] required frequency resolution - CW analysis
T = 1/DF # [s] sampling duration - CW analysis (5 mSec?)
t = np.arange(0, T, 1/Fs)
N = t.size              # length of signal
N2 = int(np.ceil(N/2))  # half length
freqs = fft.fftfreq(N, 1/Fs)[:N2]    # Fourier frequencies
sigma = 0.01
mean = 0.0
noise = np.random.normal(mean, sigma, N)
phase = 0.0
S1 = np.sin(_2Pi_*f*t) # transmit signal
S2 = np.sin(_2Pi_*f2*t + phase) # receive signal - Doppler shifted (without noise)
```

```python

start = 249000
gate = 100
stop = start + gate
plt.plot(t[start:stop]*1000, S1[start:stop], '-.', t[start:stop]*1000, S2[start:stop])
plt.legend(['pulse', 'echo'])
plt.xlabel("[mSec.]")
plt.grid()
plt.title("The Pulse and the Doppler shifted Echo - time domain");
```

```python
# Fourier Transforms
S1_ = fft.fft(S1)[:N2]
S2_ = fft.fft(S2)[:N2]
S1_max = np.abs(S1_).argmax()
freqs[S1_max]/1e3
```

```python
plt.subplot(2,1,1)
plt.plot(freqs/1e6, np.log(np.abs(S1_)), label = 'pulse')
plt.plot(freqs/1e6, np.log(np.abs(S2_)), label = 'echo')
plt.grid()
plt.legend()
#plt.xlabel('[MHz]')
plt.title('The spectrum of the pulse and the Doppler shifted echo')
plt.subplot(2,1,2)
plt.plot(freqs/1e6, np.log(np.abs(S1_)), label = 'pulse')
plt.plot(freqs/1e6, np.log(np.abs(S2_)), label = 'echo')
plt.grid()
plt.legend()
plt.xlabel('[MHz]')
window = 0.12
start = freqs[S1_max]/1e6 - window/2
stop = freqs[S1_max]/1e6 + window/2
plt.xlim([start, stop])
```

```python
# Standard processing
#
# Demodulation
S2D = S2*S1
S2D_ = fft.fft(S2D)[:N2]
# Filter
b, a = signal.butter(5, 20e+3, fs=Fs, btype = 'low', analog = False)
S2DF = signal.lfilter(b, a, S2D)
S2DF_ = fft.fft(S2DF)[:N2]
```

```python
plt.plot(freqs/1e3, np.log(np.abs(S2D_)), label = 'demodulated echo')
plt.plot(freqs/1e3, np.log(np.abs(S2DF_)), label = 'demodulated and filtered echo')
plt.grid()
plt.title('The spectrum of the echo after demodulation and LP filter')
plt.xlabel('[kHz]')
plt.legend()
plt.xlim([0, 20])
plt.xticks(np.arange(0,22,2));
```

## The Doppler shift from an oscillating scatterer - the CW analysis

```python
# Oscillating scatterer - model assumptions
F = 1.0e6 # [Hz] the particle frequency
p = 1.5e6 # [Pa] acoustic pressure amplitude
U = p/Z # [m/s] acoustic particle velocity amplitude
phase = 0.0
u = U*np.sin(_2Pi_*F*t + phase) # acoustic particle velocity
fd = 2*f*u/c # time-dependent Doppler shift 
f2 = f + fd # The frequency is modulated at the frequency of the oscillating particle F
S2 = np.sin(_2Pi_*f2*t + phase) # receive signal - Time Dependent Doppler Shift (without noise)
```

```python
# Processing - FFT
S2_ = fft.fft(S2)[:N2]
plt.plot(freqs/1e6, np.log(np.abs(S1_)), label = "original pulse")
plt.plot(freqs/1e6, np.log(np.abs(S2_)), label = "Doppler shifted echo")
plt.grid()
plt.legend()
#plt.axvline(x = f/1e6, ls = '--', color = 'r', lw = 1.0)
plt.xlabel("[MHz]")
plt.title('The spectrum of the oscillatory Doppler shift')
```

```python
# Processing - demodulation (the standrad version)
S2D = S2*S1
S2D_ = fft.fft(S2D)[:N2]
```

```python
# Processing - low pass filter
b, a = signal.butter(5, 20e+3, fs=Fs, btype = 'low', analog = False)
S2DF = signal.lfilter(b, a, S2D)
S2DF_ = fft.fft(S2DF)[:N2]
# measurement - finding the last peak in the band (peak of highest frequency)
argf = int((freqs[:100]*np.abs(S2DF_)[:100]).argmax())
maxf = freqs[argf]
maxf
```

```python
plt.plot(freqs[:100]/1e3, np.log(np.abs(S2DF_[:100])), label = 'spectrum')
plt.xticks(np.arange(0,22,2))
plt.grid()
plt.xlabel("[kHz]")
plt.axvline(x = 2*f*U/c/1e3, ls = '--', color = 'r', lw = 0.75, label = "Reference")
plt.axvline(x = maxf/1e3, ls = ':', color = 'b', lw = 1.0, label = "Measurement")
plt.title("Oscillatory Doppler shift \n the spectrum of the demodulated filtered echo")
plt.legend()
```

```python
# Unconventional demodulation - Self Demodulation
samples_in_half_period = int(Fs/F/2)
samples_in_half_period
```

```python
def my_shift(x, steps=1):
    l1 = x.tolist()
    for i in range(steps): l1.append(l1.pop(0))
    return np.asarray(l1)
```

```python
# cyclic shifting the echo signal in 1/2 period of the particle oscillation
S2shift = my_shift(S2, samples_in_half_period)
# demodulating the echo with a shifted echo - self demod
S2D = S2*S2shift
S2D_ = fft.fft(S2D)[:N2]
# Filter
b, a = signal.butter(5, 40e+3, fs=Fs, btype = 'low', analog = False)
S2DF = signal.lfilter(b, a, S2D)
# FFT
S2DF_ = fft.fft(S2DF)[:N2]
# measurement - the frequency of the last peak in the band
argf = (freqs[:100]*np.abs(S2DF_)[:100]).argmax()
maxf = freqs[argf]
maxf
```

```python
plt.plot(freqs[:200]/1e3, np.log(np.abs(S2DF_[:200])), label = 'spectrum')
plt.xticks(np.arange(0,42,2))
plt.grid()
plt.xlabel("[kHz]")
plt.axvline(x = 2*2*f*U/c/1e3, ls = '--', color = 'r', lw = 0.75, label = "Reference")
plt.axvline(x = maxf/1e3, ls = ':', color = 'b', lw = 1.0, label = "Measurement")
plt.title("Oscillatory Doppler shift \n the spectrum of the self-demodulated filtered echo")
plt.legend()
```
